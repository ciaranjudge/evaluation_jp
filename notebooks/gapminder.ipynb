```python
# create a new plot with default tools, using figure
p = figure(plot_width=600, plot_height=600)
p.x_range = Range1d(6000, 15000)
p.y_range = Range1d(0, 9000)
p.xaxis.axis_label = "Social Welfare income"
p.yaxis.axis_label = "Employment earnings"
# color_mapper = CategoricalColorMapper(palette=Spectral6, factors=data["cluster name"])

# add a circle renderer with x and y coordinates, size, color, and alpha
ordered_cluster_names = [
    "Younger  Professionals",
    "Shorter-Term Unemployed",
    "Younger  Casual Claimants",
    "Intermittent Labour Market Attachment",
    "Older, with Strong Employment History ",
    "Self-Employed",
    "Longer-Term Unemployed",
]

label = Label(
    x=6010,
    y=10,
    text=str(years[0]),
    text_font_size="70pt",
    text_color="#eeeeee",
    # level=0,
)


p.circle(
    x="w_sw_pay_",
    y="w_Earn",
    source=source_c,
    radius="clust_size",
    fill_color=factor_cmap(
        "cluster name", palette=Spectral6, factors=ordered_cluster_names
    ),
    alpha=0.3,
)

p.circle(
    x="w_sw_pay_",
    y="w_Earn",
    source=source_t,
    radius="clust_size",
    fill_color=factor_cmap(
        "cluster name", palette=Spectral6, factors=ordered_cluster_names
    ),
    # legend="cluster name",
    alpha=0.9,
)

labels_t = LabelSet(
    x="w_sw_pay_",
    y="w_Earn",
    source=source_t,
    text="cluster",
    level="glyph",
    text_align="center",
    y_offset=5,
    render_mode="canvas",
    text_font_size="8pt",
)
labels_c = LabelSet(
    x="w_sw_pay_",
    y="w_Earn",
    source=source_c,
    text="cluster",
    level="glyph",
    text_align="center",
    y_offset=5,
    render_mode="canvas",
    text_font_size="8pt",
    text_alpha=0.5,
)
p.add_layout(label)
p.add_layout(labels_t)
p.add_layout(labels_c)

def animate_update():
    year = slider.value + 1
    if year > years[-1]:
        year = years[0]
    slider.value = year


def slider_update(attrname, old, new):
    year = slider.value
    label.text = str(year)
    source_t.data = year_ts[year]
    source_c.data = year_cs[year]


slider = Slider(
    start=years[0], end=years[-1], value=years[0], step=1, title="Year"
)
slider.on_change("value", slider_update)

callback_id = None


def animate():
    global callback_id
    if button.label == " Play":
        button.label = " Pause"
        callback_id = curdoc().add_periodic_callback(animate_update, 200)
    else:
        button.label = " Play"
        curdoc().remove_periodic_callback(callback_id)


button = Button(label=" Play", width=60)
button.on_click(animate)

layout = layout([[p], [slider, button]], sizing_mode="scale_width")

curdoc().add_root(layout)
curdoc().title = "Gapminder"
show(layout)  # show the results

```
